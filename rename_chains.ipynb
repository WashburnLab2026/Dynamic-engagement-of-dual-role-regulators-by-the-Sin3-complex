```python
""" 
Proposed by Jules Nde, PhD, Washburn Lab, Cancer Biology, University of Kansas Medical Center (25/8/2025)

Author: Jules Nde, Washburn Lab, Cancer Biology, University of Kansas Medical Center
Copyright © 2025 by Jules Nde. All rights reserved. No part of this work may be reproduced or transmitted in any form without permission.

This script re-annotates one or more chains in a given PDB file and writes the result to a new PDB file.
e.g. of usage: $python reannotate_chains.py input.pdb output.pdb B,C A   (if using the python version instead of jupyter notebook)
This command will rename chains B and C from the input.pdb to A and save the result to the output.pdb
"""

from Bio.PDB import PDBParser, PDBIO, Select
import os
# Cell 2: ChainRenamer Class
class ChainRenamer(Select):
    def __init__(self, source_chains, target_chain_id):
        self.source_chains = source_chains
        self.target_chain_id = target_chain_id

    def accept_chain(self, chain):
        return True  # Accept all chains

    def accept_residue(self, residue):
        return True

    def accept_atom(self, atom):
        return True

    def rename_chain(self, model):
        for chain in model:
            if chain.id in self.source_chains:
                chain.id = self.target_chain_id

def reannotate_pdb_chains(input_pdb, output_pdb, source_chains, target_chain_id):
    parser = PDBParser(QUIET=True)
    structure = parser.get_structure("structure", input_pdb)

    
    for model in structure:
        ChainRenamer(source_chains, target_chain_id).rename_chain(model)

    
    io = PDBIO()
    io.set_structure(structure)
    io.save(output_pdb)

# Provide your input values here
input_pdb = "cluster8_4_Rpd3L_Cti6-Core_6XAW25A_Sin3FFI.pdb"  # Replace with the path to your input PDB file
output_pdb = "cluster8_4_Rpd3L_Cti6-Core_6XAW25A_Sin3FFI_Q.pdb"  # Desired output filename
#source_chains = ["B", "C", "D", "E", "F", "G"]  # List of chain IDs to rename
source_chains = ["B"]  # List of chain IDs to rename
target_chain = "Q"          # Target chain ID

reannotate_pdb_chains(input_pdb, output_pdb, source_chains, target_chain)
print(f"Renamed chains {source_chains} to '{target_chain}' in '{output_pdb}'")
```
